### **Attributes' histograms**

```python
from src.data.labels import labels

labels_idx: dict = {label: i for i, label in enumerate(labels) if i != 0}

df = fixed_dataset.data[fixed_dataset.data["Label"] != 0]
```

```python
import matplotlib.pyplot as plt
import numpy as np


total_steps = len(df.columns)

for step, column in enumerate(df.columns):
    progress = (step + 1) / total_steps
    progress = (step + 1) / total_steps
    progress_bar = (
        "[" + "=" * int(50 * progress) + " " * (50 - int(50 * progress)) + "]"
    )
    print(f"\r{progress_bar} {step + 1}/{total_steps} attributes processed", end="")
    if np.issubdtype(df[column].dtype, np.number):
        try:
            mask = np.isfinite(df[column])
            data = df[column][mask]
            # bins = histogram_bin_edges(data, "auto")
            bins = 500

            plt.figure(figsize=(10, 6))  # Ustaw rozmiar wykresu
            plt.hist(data, bins)
            plt.title(f"Rozkład atrybutu: {column}")  # Ustaw tytuł wykresu
            plt.xlabel(column)  # Ustaw etykietę osi X
            plt.ylabel("Liczba")  # Ustaw etykietę osi Y
            plt.savefig(f"./img/histograms/CICIDS2018/{column}_rozklad.png")  # Save the plot as a PNG file
            plt.close()
            plt.clf()
        except:
            print(f"\nError with column: {column}")
            plt.close()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = fixed_dataset.data.corr()
label_corr = corr[["Label"]].drop(["Label"], axis=0)
n = 11

num_attributes = label_corr.shape[0]
attributes_per_subplot = num_attributes // n

fig, axes = plt.subplots(nrows=1, ncols=n, figsize=(n * 3, 16))
cmap = "rocket_r"

for i in range(n):
    start_idx = i * attributes_per_subplot
    end_idx = (i + 1) * attributes_per_subplot
    subset = label_corr[start_idx:end_idx]

    sns.heatmap(
        subset,
        cmap=cmap,
        annot=False,
        vmax=1,
        square=True,
        center=0,
        cbar_kws={"shrink": 0.2},
        ax=axes[i],
    )
    axes[i].set_title(f"Attributes {start_idx+1} to {end_idx}")
    if i == 0:
        axes[i].set_ylabel("Attributes")
    axes[i].tick_params(axis="y", labelrotation=0)

plt.tight_layout(pad=1.0)

plt.savefig(f"./img/heatmaps/ids2018_label_atrybuty_multicol.png", bbox_inches="tight")
plt.clf()
```

### **Confusion matrix**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


def plot_confusion_matrix(values, percentage=False):
    PP, FP, FN, TN = values["PP"], values["FP"], values["FN"], values["TN"]
    total = PP + FP + FN + TN

    matrix = np.array([[PP, FP], [FN, TN]])
    fmt = ".0f"

    if percentage:
        matrix = matrix / total * 100
        fmt = ".2f"

    labels = ["Pozytywny", "Negatywny"]

    plt.figure(figsize=(10, 7))
    sns.heatmap(
        matrix,
        annot=True,
        fmt=fmt,
        cmap="rocket_r",
        xticklabels=labels,
        yticklabels=labels,
    )

    plt.title("Macierz konfuzji" + (" (Procentowo)" if percentage else " (Liczbowo)"))

    plt.show()


values = {"PP": 1075889, "FP": 154260, "FN": 434342, "TN": 2162578}

plot_confusion_matrix(values, percentage=False)
plot_confusion_matrix(values, percentage=True)
```
