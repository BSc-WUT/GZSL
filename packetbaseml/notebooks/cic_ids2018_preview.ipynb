### **Attributes' histograms**

```python
from src.data.labels import labels

labels_idx: dict = {label: i for i, label in enumerate(labels) if i != 0}

df = fixed_dataset.data[fixed_dataset.data["Label"] != 0]
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import histogram_bin_edges


total_steps = len(df.columns)

for step, column in enumerate(df.columns):
    progress = (step + 1) / total_steps
    progress = (step + 1) / total_steps
    progress_bar = (
        "[" + "=" * int(50 * progress) + " " * (50 - int(50 * progress)) + "]"
    )
    print(f"\r{progress_bar} {step + 1}/{total_steps} attributes processed", end="")
    if np.issubdtype(df[column].dtype, np.number):
        try:
            mask = np.isfinite(df[column])
            data = df[column][mask]
            # bins = histogram_bin_edges(data, "auto")
            bins = 500

            plt.figure(figsize=(10, 6))  # Ustaw rozmiar wykresu
            plt.hist(data, bins)
            plt.title(f"Rozkład atrybutu: {column}")  # Ustaw tytuł wykresu
            plt.xlabel(column)  # Ustaw etykietę osi X
            plt.ylabel("Liczba")  # Ustaw etykietę osi Y
            plt.savefig(f"./img/histograms/CICIDS2018/{column}_rozklad.png")  # Save the plot as a PNG file
            plt.close()
            plt.clf()
        except:
            print(f"\nError with column: {column}")
            plt.close()
```

### **Attributes' Heatmap**

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = fixed_dataset.data.corr()

label_corr = corr[["Label"]].drop(["Label"], axis=0)  

num_attributes = label_corr.shape[0]
attributes_per_subplot = num_attributes // 4  


fig, axes = plt.subplots(
    nrows=1, ncols=4, figsize=(12, 20) 
)

cmap = sns.diverging_palette(230, 20, as_cmap=True)

for i in range(4):
    start_idx = i * attributes_per_subplot
    end_idx = (i + 1) * attributes_per_subplot
    subset = label_corr[start_idx:end_idx]

    # Create the heatmap for the current subset
    sns.heatmap(
        subset,
        cmap=cmap,
        annot=False,  # Include annotations
        vmax=0.3,
        center=0,
        square=True,  # Set to True to make cells square
        linewidths=0.2,
        cbar_kws={"shrink": 0.5},
        ax=axes[i],  # Plot in the ith subplot
    )
    axes[i].set_title("Korelacja między Label a pozostałymi atrybutami")
    axes[i].set_xlabel("Label")
    axes[i].set_ylabel("Atrybuty")
    axes[i].tick_params(axis="y", labelsize=8)  # Adjust label size for readability

# Adjust the layout
plt.tight_layout()

# Save the plot
plt.savefig(f"./img/heatmaps/ids2018_label_atrybuty_multicol.png", bbox_inches="tight")
plt.clf()
```

```python
columns_to_drop = corr[(corr["Label"] >= -0.05) & (corr["Label"] <= 0.05)].index

# Print the columns to be dropped
print("Columns to be dropped due to low correlation with 'Label':")
print(columns_to_drop)
```

```python
fixed_dataset.data.drop(columns=columns_to_drop, inplace=True)
```
